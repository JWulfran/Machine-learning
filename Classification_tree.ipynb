```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
np.bincount(iris.target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
model_tree = tree.DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
# Visualisation de l'arbre 
plt.figure(figsize = (10,8))
plot_tree(model_tree)
plt.show()
```

```python
accuracy_tree_test = model_tree.score(X_test,y_test)
print(accuracy_tree_test)
```

```python
mdepth = []
msplit = []

for i in range(2,10):
    clt = tree.DecisionTreeClassifier(random_state=42, max_depth=i)
    clt = clt.fit(X_train,y_train)
    mdepth.append(clt.score(X_test,y_test))

for j in range(2,10):
    clt = tree.DecisionTreeClassifier(random_state=42, min_samples_split=j)
    clt = clt.fit(X_train, y_train)
    msplit.append(clt.score(X_test, y_test))
```

```python
print(mdepth)
print(msplit)
```

```python
plt.scatter(mdepth, msplit)
plt.show()
```
